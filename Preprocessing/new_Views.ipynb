```python
df = pd.read_csv("C:/Users/SOYOUNG/Desktop/SOYOUNG/daitgirls/team project/TED/ted-talks/ted_main.csv")
df.columns
```

```python
df.head()
```

```python
df.loc[:, 'new_column'] = time.mktime(d.timetuple())
```

```python
df.head()
```

```python
df['date_ratio'] = df['published_date'] / df['new_column'] 
df
```

```python
df['new_views'] = df['date_ratio'] * df['views']
df
```

```python
df['published_date'] = df['published_date'].apply(lambda x: datetime.datetime.fromtimestamp( int(x)).strftime('%d-%m-%Y'))
```

```python
df.iloc[1]['ratings']
```

```python
import ast
df['ratings'] = df['ratings'].apply(lambda x: ast.literal_eval(x))
```

```python
df['funny'] = df['ratings'].apply(lambda x: x[0]['count'])
df['coureagous'] = df['ratings'].apply(lambda x: x[1]['count'])
df['confusing'] = df['ratings'].apply(lambda x: x[2]['count'])
df['beautiful'] = df['ratings'].apply(lambda x: x[3]['count'])
df['unconvincing'] = df['ratings'].apply(lambda x: x[4]['count'])
df['longwinded'] = df['ratings'].apply(lambda x: x[5]['count'])
df['informative'] = df['ratings'].apply(lambda x: x[6]['count'])
df['inspiring'] = df['ratings'].apply(lambda x: x[7]['count'])
df['fascinating'] = df['ratings'].apply(lambda x: x[8]['count'])
df['ingenious'] = df['ratings'].apply(lambda x: x[9]['count'])
df['persuasive'] = df['ratings'].apply(lambda x: x[10]['count'])
df['jawdrop'] = df['ratings'].apply(lambda x: x[-3]['count'])
df['obnoxious'] = df['ratings'].apply(lambda x: x[-2]['count'])
df['OK'] = df['ratings'].apply(lambda x: x[-1]['count'])

df.head()
```

```python
pop_talk = df[['title', 'main_speaker', 'new_views', 'published_date',  'description',  'speaker_occupation', 
                'num_speaker', 'duration', 'comments', 'tags', 'languages', 'url','funny', 'coureagous', 'confusing','beautiful', 'unconvincing',
              'longwinded', 'informative', 'inspiring', 'fascinating', 'ingenious', 'persuasive', 'jawdrop', 'obnoxious','OK' ]].sort_values('new_views', ascending=False)[:]
pop_talk
```

```python
pop_talk.to_csv('ted_newviews', sep='\t', encoding='utf-8')
```
